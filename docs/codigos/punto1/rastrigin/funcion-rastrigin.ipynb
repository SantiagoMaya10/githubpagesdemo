# **Función de Rastrigin**

### **Características e historia:**
La función de Rastrigin es una función no convexa usada como una función para probar el rendimiento en algoritmos de optimización. Fue propuesta por Rastrigin como una función bidimensional en 1974 y, fue generalizada por Rudolph. Se popularizó debido a estudios de Hoffmeister y Bäck ("Genetic Algorithms and Evolution Strategies: Similarities and Differences"), también de
Mühlenbein et al ("The Parallel Genetic Algorithm as Function Optimizer ").

### **Principal característica**
Hallar el mínimo global de esta función es un problema complejo por su gran espacio de búsqueda y el gran número de mínimos locales.


### **Definición N-Dimensional**


$f(x) = 10d +\sum_{i=1}^{d}[x_{i}^2 - 10\cos(2 \pi x_{i})]$

La función es usualmente evaluada en el hipercubo $ x_{i} \in [-5.12, 5.12] \space \forall \space i=1,.., d. $


### **Definición en 2 dimensiones**

$f(x_{1}, x_{2}) = 20 + x_{1}^2 + x_{2}^2 - 10(\cos(2 \pi x_{1}) + \cos(2 \pi x_{2}))$



### **Definición en 3 dimensiones**

$f(x_{1}, x_{2}, x_{3}) = 30 + x_{1}^2 + x_{2}^2 + x_{3}^2 - 10(\cos(2 \pi x_{1}) + \cos(2 \pi x_{2}) + \cos(2 \pi x_{3}))$



### **Gráfico de Rastrigin en dos dimensiones**



```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

# Definir la función de dos variables
def rastrigin_function(x1, x2):
    return 20 + (x1)**2 + (x2)**2 - 10 * (np.cos(2*np.pi*x1) + np.cos(2*np.pi*x2))

# Generar datos para los ejes x e y
x = np.linspace(-5.12, 5.12, 1000)
y = np.linspace(-5.12, 5.12, 1000)
x, y = np.meshgrid(x, y)
z = rastrigin_function(x, y)

# Crear la figura y los ejes 3D
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')

# Graficar el plano en 3D
ax.plot_surface(x, y, z, cmap='viridis')

# Etiquetas de los ejes
ax.set_xlabel('X1')
ax.set_ylabel('X2')
ax.set_zlabel('X3')

# Rotar el gráfico
ax.view_init(elev=10, azim=100)

# Punto de interés
x_point = 0
y_point = 0
z_point = rastrigin_function(x_point, y_point)

# Añadir un punto en la superficie
ax.scatter(x_point, y_point, z_point, color='red', s=70)

ax.text(x_point, y_point + 1.5, z_point, "(0, 0, 0)", color='black')

plt.title('Función de Ragistrin en 2 dimensiones')


# Mostrar el gráfico
plt.show()
```

La siguiente gráfica muestra las curvas de nivel de la función en dos dimensiones con 100 curvas de nivel:

```python
# Crear el gráfico con 100 curvas de nivel
plt.contour(x, y, z, levels=100, cmap='viridis')

# Etiquetas de los ejes
plt.xlabel('X1')
plt.ylabel('X2')

plt.scatter(0, 0, color='red', label='(0, 0)')

# Labels for the axes
plt.xlabel('X1')
plt.ylabel('X2')

# Mostrar el gráfico
plt.colorbar(label='Niveles')
plt.title('Gráfica con 100 curvas de nivel de la función de Rastrigin $f(x_{1},x_{2})$')
plt.grid(True)
plt.xlim(-1,1)
plt.ylim(-1,1)
plt.show()
```

Aquí podemos observar que el punto mínimo global se encuentra en (0,0) y con solo 100 curvas está bien acotado. Pero veamoslo con 1000 curvas de nivel:

```python
# Crear el gráfico con 1000 curvas de nivel
plt.contour(x, y, z, levels=1000, cmap='viridis')

# Etiquetas de los ejes
plt.xlabel('X1')
plt.ylabel('X2')

plt.scatter(0, 0, color='red', label='(0, 0)')

# Labels for the axes
plt.xlabel('X1')
plt.ylabel('X2')

# Mostrar el gráfico
plt.colorbar(label='Niveles')
plt.title('Gráfica con 1000 curvas de nivel de la función de Rastrigin $f(x_{1},x_{2})$')
plt.grid(True)
plt.xlim(-0.25,0.25)
plt.ylim(-0.25,0.25)
plt.show()
```

Se ilustran alrededor varias puntos mínimos locales también encerrados por curvas de nivel, esto permite ver la dificultad de encontrar el mínimo de esta función. Sin embargo, se acota en el mínimo global señalado en rojo que es (0,0)
